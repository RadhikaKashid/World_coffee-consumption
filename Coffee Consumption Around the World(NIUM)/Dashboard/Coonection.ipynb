```python
import streamlit as st
import pandas as pd
from sqlalchemy import create_engine
import psycopg2 as pg



DB_USER = "postgres"
DB_PASSWORD = "1234"
DB_HOST = "localhost"
DB_PORT = "5432"
DB_NAME = "postgres"

engine = create_engine(
    f"postgresql://{DB_USER}:{DB_PASSWORD}@{DB_HOST}:{DB_PORT}/{DB_NAME}"
)

try:
    query = 'SELECT * FROM "Nium".country_bridge LIMIT 5;'

    df = pd.read_sql(query, engine)

    print("✅ Connected Successfully")
    print(df)

except Exception as e:
    print("❌ Connection Failed")
    print(e)
```

```python
import psycopg2

try:
    conn = psycopg2.connect(
        host="localhost",
        database="postgres",
        user="postgres",
        password="1234",
        port="5432"
    )

    cursor = conn.cursor()

    cursor.execute("""
        SELECT *
        FROM "Nium".country_bridge
        LIMIT 5;
    """)

    rows = cursor.fetchall()

    print("✅ Connected Successfully")

    for row in rows:
        print(row)

    cursor.close()
    conn.close()

except Exception as e:
    print("❌ Connection Failed")
    print(e)
```
