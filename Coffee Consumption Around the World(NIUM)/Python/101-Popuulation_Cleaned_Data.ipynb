```python
import pandas as pd
import numpy as np
```

```python

# File path
input_file = r"C:\Users\Radhika Kashid\Downloads\psd_coffee_csv\Poppulation.csv"   # Update your path
output_file = r"C:\Users\Radhika Kashid\Downloads\Cleaned Data\Cleaned_Poppulation.csv"

# Read data while skipping metadata rows
df = pd.read_csv(
    input_file,
    skiprows=4
)

# Remove completely empty columns
df = df.dropna(axis=1, how='all')

# Remove completely empty rows
df = df.dropna(how='all')

# Remove duplicate rows
df = df.drop_duplicates()

# Clean column names
df.columns = (
    df.columns
      .str.strip()
      .str.replace(" ", "_")
      .str.replace(r"[^A-Za-z0-9_]", "", regex=True)
)

# Replace empty strings with NaN
df.replace("", np.nan, inplace=True)

# Convert year columns to numeric
year_cols = [col for col in df.columns if col.isdigit()]

for col in year_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Remove rows where Country Name is missing
df = df.dropna(subset=['Country_Name'])

# Rename columns for readability
df.rename(columns={
    'Country_Name': 'Country',
    'Country_Code': 'Country_Code',
    'Indicator_Name': 'Indicator_Name',
    'Indicator_Code': 'Indicator_Code'
}, inplace=True)

# Save cleaned file
df.to_csv(output_file, index=False)

print("Data Cleaning Completed Successfully")
print(f"Shape of cleaned dataset: {df.shape}")
print(f"Output saved at: {output_file}")

# Preview
display(df.head())
```

```python
# Convert wide format to long format

id_cols = ['Country', 'Country_Code', 'Indicator_Name', 'Indicator_Code']

df_long = df.melt(
    id_vars=id_cols,
    value_vars=year_cols,
    var_name='Year',
    value_name='Population'
)

df_long['Year'] = df_long['Year'].astype(int)

df_long.to_csv("Population_Long_Format.csv", index=False)

print(df_long.head())
```
