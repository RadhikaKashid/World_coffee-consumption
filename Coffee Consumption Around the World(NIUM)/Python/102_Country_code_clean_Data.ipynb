```python
import pandas as pd
import os

# =====================================================
# INPUT AND OUTPUT FILE PATHS
# =====================================================

input_file = r"C:\Users\Radhika Kashid\Downloads\psd_coffee_csv\countries-codes.csv"
output_file = r"C:\Users\Radhika Kashid\Downloads\Cleaned Data\countries-codes-cleaned.csv"

# Create output folder if it doesn't exist
os.makedirs(os.path.dirname(output_file), exist_ok=True)

# =====================================================
# LOAD DATA
# =====================================================

df = pd.read_csv(input_file)

print(f"Original Rows    : {len(df)}")
print(f"Original Columns : {len(df.columns)}")

# =====================================================
# STANDARDIZE COLUMN NAMES
# =====================================================

df.columns = (
    df.columns
      .str.strip()
      .str.upper()
      .str.replace(" ", "_")
      .str.replace("-", "_")
)

# =====================================================
# REMOVE DUPLICATES
# =====================================================

df = df.drop_duplicates()

# =====================================================
# REMOVE COMPLETELY EMPTY ROWS
# =====================================================

df = df.dropna(how='all')

# =====================================================
# CLEAN TEXT COLUMNS
# =====================================================

text_columns = df.select_dtypes(include='object').columns

for col in text_columns:
    df[col] = (
        df[col]
        .astype(str)
        .str.strip()
        .str.replace(r'\s+', ' ', regex=True)
    )

# =====================================================
# REPLACE BLANK VALUES WITH NULL
# =====================================================

df.replace(r'^\s*$', pd.NA, regex=True, inplace=True)

# =====================================================
# REMOVE LARGE GEO SHAPE COLUMN (OPTIONAL)
# =====================================================

columns_to_drop = ['GEO_SHAPE']

existing_cols = [c for c in columns_to_drop if c in df.columns]

if existing_cols:
    df.drop(columns=existing_cols, inplace=True)

# =====================================================
# CLEAN BOOLEAN COLUMNS
# =====================================================

bool_cols = ['IS_ILOMEMBER', 'IS_RECEIVING_QUEST']

for col in bool_cols:
    if col in df.columns:
        df[col] = df[col].astype(str).str.upper().str.strip()

# =====================================================
# RESET INDEX
# =====================================================

df.reset_index(drop=True, inplace=True)

# =====================================================
# SAVE CLEANED DATA
# =====================================================

df.to_csv(output_file, index=False)

print("\nCleaning Completed Successfully")
print(f"Cleaned Rows    : {len(df)}")
print(f"Cleaned Columns : {len(df.columns)}")
print(f"Output File     : {output_file}")
```

```python
import pandas as pd

# =====================================================
# INPUT & OUTPUT FILES
# =====================================================

input_file = r"C:\Users\Radhika Kashid\Downloads\Cleaned Data\countries-codes-cleaned_v2.csv"
output_file = r"C:\Users\Radhika Kashid\Downloads\Cleaned Data\countries-codes-cleaned_v3.csv"

# =====================================================
# READ FILE
# =====================================================

df = pd.read_csv(input_file)

# =====================================================
# SPLIT GEO_POINT_2D COLUMN
# =====================================================

if 'GEO_POINT_2D' in df.columns:

    # Split into two columns
    df[['GEO_POINT_X', 'GEO_POINT_Y']] = (
        df['GEO_POINT_2D']
        .astype(str)
        .str.split(',', expand=True)
    )

    # Remove leading/trailing spaces
    df['GEO_POINT_X'] = df['GEO_POINT_X'].str.strip()
    df['GEO_POINT_Y'] = df['GEO_POINT_Y'].str.strip()

    # Optional: Convert to numeric
    df['GEO_POINT_X'] = pd.to_numeric(df['GEO_POINT_X'], errors='coerce')
    df['GEO_POINT_Y'] = pd.to_numeric(df['GEO_POINT_Y'], errors='coerce')

else:
    print("Column GEO_POINT_2D not found.")

# =====================================================
# SAVE NEW FILE
# =====================================================

df.to_csv(output_file, index=False)

print("File processed successfully.")
print(f"Output saved to:\n{output_file}")
```
