```python
import pandas as pd
import os

# =====================================================
# INPUT FILE
# =====================================================
input_file = r"C:\Users\Radhika Kashid\Downloads\Cleaned Data\Cleaned_Poppulation.csv"

# =====================================================
# OUTPUT FOLDER
# =====================================================
input_folder = os.path.dirname(input_file)

output_folder = os.path.join(input_folder, "Fact_Population")
os.makedirs(output_folder, exist_ok=True)

# =====================================================
# OUTPUT FILE
# =====================================================
output_file = os.path.join(
    output_folder,
    "Cleaned_Poppulation_Fact.csv"
)

# =====================================================
# READ INPUT FILE
# =====================================================
print("Reading file...")
df = pd.read_csv(input_file)

print(f"Original Shape: {df.shape}")

# =====================================================
# IDENTIFY YEAR COLUMNS
# =====================================================
year_columns = [col for col in df.columns if str(col).isdigit()]

print(f"Year Columns Found: {len(year_columns)}")

# =====================================================
# UNPIVOT / MELT DATA
# =====================================================
fact_df = pd.melt(
    df,
    id_vars=[
        'Country',
        'Country_Code',
        'Indicator_Name',
        'Indicator_Code'
    ],
    value_vars=year_columns,
    var_name='year',
    value_name='value'
)

# =====================================================
# DATA TYPE CONVERSION
# =====================================================
fact_df['year'] = pd.to_numeric(
    fact_df['year'],
    errors='coerce'
).astype('Int64')

fact_df['value'] = pd.to_numeric(
    fact_df['value'],
    errors='coerce'
).astype('Int64')

# =====================================================
# REMOVE NULL VALUES (OPTIONAL)
# =====================================================
fact_df = fact_df.dropna(subset=['value'])

# =====================================================
# RENAME COLUMNS TO MATCH FACT TABLE
# =====================================================
fact_df = fact_df.rename(columns={
    'Country': 'country',
    'Country_Code': 'country_code',
    'Indicator_Name': 'indicator_name',
    'Indicator_Code': 'indicator_code'
})

# =====================================================
# ADD ID COLUMN (OPTIONAL)
# =====================================================
fact_df.insert(0, 'id', range(1, len(fact_df) + 1))

# =====================================================
# SAVE OUTPUT
# =====================================================
fact_df.to_csv(output_file, index=False)

print("\nTransformation Completed Successfully")
print(f"Output Shape: {fact_df.shape}")
print(f"Output File: {output_file}")

# Preview
print("\nSample Data:")
print(fact_df.head())
```
