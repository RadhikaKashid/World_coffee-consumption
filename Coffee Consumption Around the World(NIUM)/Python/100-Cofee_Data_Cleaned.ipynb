```python
import pandas as pd

# ==========================================
# Specify file paths
# ==========================================

input_file = r"C:\Users\Radhika Kashid\Downloads\psd_coffee_csv\psd_Coffee.csv"
output_file = r"C:\Users\Radhika Kashid\Downloads\Cleaned Data\psd_coffee_cleaned.csv"

# ==========================================
# Load data
# ==========================================

df = pd.read_csv(input_file)

print(f"Original Records: {len(df)}")
print(f"Original Columns: {len(df.columns)}")

# ==========================================
# Remove duplicate rows
# ==========================================

df.drop_duplicates(inplace=True)

# ==========================================
# Remove rows with all null values
# ==========================================

df.dropna(how='all', inplace=True)

# ==========================================
# Trim spaces from text columns
# ==========================================

for col in df.select_dtypes(include=['object']).columns:
    df[col] = df[col].astype(str).str.strip()

# ==========================================
# Replace blank strings with NaN
# ==========================================

df.replace(r'^\s*$', pd.NA, regex=True, inplace=True)

# ==========================================
# Remove rows where all values are null
# ==========================================

df.dropna(how='all', inplace=True)

# ==========================================
# Convert numeric columns if present
# ==========================================

numeric_columns = [
    'Commodity_Code',
    'Market_Year',
    'Calendar_Year',
    'Month',
    'Attribute_ID',
    'Unit_ID',
    'Value'
]

for col in numeric_columns:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')

# ==========================================
# Remove rows with null Value
# ==========================================

if 'Value' in df.columns:
    df = df[df['Value'].notna()]

# ==========================================
# Reset Index
# ==========================================

df.reset_index(drop=True, inplace=True)

# ==========================================
# Save cleaned data
# ==========================================

df.to_csv(output_file, index=False)

print(f"Cleaned file saved successfully.")
print(f"Location: {output_file}")
print(f"Final Records: {len(df)}")
```
